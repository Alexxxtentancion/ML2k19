```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_wine
```

```python
wine_dataset = load_wine()
print(wine_dataset.keys())
```

```python
print('Описание: \n{}'.format(wine_dataset['DESCR']))
```

```python
target_names = pd.DataFrame({'Классы':wine_dataset['target_names']})
display(target_names)
```

```python
feature_names = pd.DataFrame({'Признаки':wine_dataset['feature_names']})
display(feature_names)
```

```python
print('Размер: \n{}'.format(wine_dataset['data'].shape))
```

```python
data = pd.DataFrame(wine_dataset['data'],columns=wine_dataset['feature_names'])
data['label'] = wine_dataset['target']
display(data)
```

```python
data.plot.scatter(x = 'alcohol', y = 'proline')
```

sns.pairplot(data)

```python
sns.violinplot(x=data['alcohol'])
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['alcohol'])
```

```python
correlation = data.corr()
plt.figure(figsize=(14, 12))
heatmap = sns.heatmap(correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
```
